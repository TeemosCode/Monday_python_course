```python
q = np.linspace(10,150,100)
w = q*np.sin(4.6*q+6) +  25*np.random.randn(100)
w_ = q*np.sin(4.6*q+6)
```

```python
plt.scatter(q,w, s = 100, alpha = 0.4)
plt.plot(q,w_,"r")
x_train, x_test, y_train, y_test = train_test_split(q,w, test_size=0.2, random_state = 1)
#defining our function
def f(x,a,b):
    return x*np.sin(a*x + b)

popt, pcov = curve_fit(f, x_train, y_train , p0=[4.6,6])
a,b = popt

plt.scatter(x_train, y_train, c = "cyan", edgecolor = "r", s = 50)
plt.scatter(x_test, y_test, c = "red", s = 40)
plt.plot(np.sort(x_test), np.sort( x_test*np.sin(a*x_test + b) ), c = "cyan", linewidth = 2, alpha = 0.2)
plt.plot(np.sort(x_train), np.sort( x_train*np.sin(a*x_train + b) ), c = "black", linewidth = 2, alpha = 0.3)
```
