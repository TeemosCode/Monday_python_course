# MNIST

```python
import keras
```

```python
from keras.datasets import mnist
```

```python
(x0_train, y0_train) , (x0_test, y0_test) = mnist.load_data()
```

```python
x_train = x0_train.reshape(60000,28*28)
```

```python
x_test = x0_test.reshape(10000, 28*28)
```

```python
from keras.utils import np_utils
```

```python
y_train = np_utils.to_categorical(y0_train, 10)
```

```python
y_test = np_utils.to_categorical(y0_test , 10)
```

```python
y_train[8787]
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense, Activation
```

```python
from keras.optimizers import SGD
```

```python
model = Sequential() # create Sequential machine
```

```python
model.add(Dense(units=500, input_dim = 784 )) #first layer
model.add(Activation("sigmoid")) #defining activation function usage
```

```python
model.add(Dense(units=500)) #second layer already learned how to look at inputs, no need to specify
model.add(Activation("sigmoid")) #defining activation function usage
```

```python
model.add(Dense(units=10)) #Output layer
model.add(Activation("softmax")) #making all vectors add up to 1
```

```python
#compile
model.compile(loss = "mse", optimizer = SGD(lr = 0.1),
             metrics=["accuracy"])
```

```python
model.fit(x_train, y_train, batch_size = 100, epochs = 20)
```
