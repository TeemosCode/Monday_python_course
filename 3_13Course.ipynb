# Supervised Learning

```python
x = np.array( [[-2,3],[-8,7],[1,-2],[2,-2]] )
y = np.array([1,1,2,2]) #Define類別   , classification
```

```python
plt.scatter(x[:,0],x[:,1], c = y, s = 50)  #How to handle a multidimensional array for drawing graph
#c = y, shows the color of x corresponding to y's elements
```

```python
x[:,1]
```

```python
from sklearn.svm import SVC
#Support vector machines (SVMs) are a set of supervised 
#learning methods used for classification, regression and outliers detection.
```

```python
clf = SVC()
```

```python
clf.fit(x,y) #Trainning data (Matching x and y)
```

```python
clf.predict([[-2,3]])  #predicts the class that corresponds to our y, should belong to class 1 in y
```

```python
clf.predict([[8,7]])
```

```python
clf.predict([[1,-2]])
```

```python
clf.predict([[5,-90]])
```

```python
# x : -10 ~ 4, y : -4 ~ 8
#grid
gd = np.array( [(i,j) for i in np.arange(-8,4,0.6) for j in np.arange(-10,6,0.8)] )
```

```python
np.arange(0,10,0.7) # the numpy arange() can let the generator skip floating points, where as the built in python
# range can only allow to jump integers value
#[(i,j) for i in np.arange(-8,4,0.6) for j in np.arange(-10,6,0.8)]
```

```python
gdc = clf.predict(gd)
print(len(gdc), len(gd[:,0]), len(gd[:,1]))
```

```python
plt.scatter(gd[:,0], gd[:,1], s = 50, c = gdc)
```

# K-Means

```python
x = np.random.rand(100,2 )# ,2 means 100 of 2 dimensional numbers between 0 to 1
```

```python
plt.scatter(x[:,0], x[:,1], s = 50)
```

```python
z = np.random.randn(100,2)
```

```python
plt.scatter(z[:,0], z[:,1], s = 60)
```

```python
from sklearn.cluster import KMeans # importing KMeans
```

```python
clf = KMeans(n_clusters=3)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c = clf.labels_ , s = 50)
```

```python
clf = KMeans(n_clusters= 10)
clf.fit(x)
clf.labels_
plt.scatter(x[:,0], x[:,1], c = clf.labels_ , s = 50)
```

```python
#Doing Learning to sort the data into groups
gd = np.array([(i,j) for i in np.arange(-0.2,1.2,0.05) for j in np.arange(-0.2,1.2,0.05)])
#gdc = clf.predict(x)
plt.scatter(gd[:,0], gd[:,1], s = 50, c = clf.predict(gd))
```

```python
#Doing Learning to sort the data into groups
gd = np.array([(i,j) for i in np.arange(2,-2,-0.35) for j in np.arange(-1,2,0.05)])
#gdc = clf.predict(x)
plt.scatter(gd[:,0], gd[:,1], s = 50, c = clf.predict(gd))
```
