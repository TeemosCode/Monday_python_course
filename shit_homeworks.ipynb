```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
```

```python
clf = SVC() # getting the svc machine
```

```python
from sklearn.datasets import make_classification
```

```python
#n_features 是指 x 的參數要幾個, n_classes 是你要分成幾類。
x, y = make_classification(n_features = 2, n_redundant = 0, n_informative = 2, n_clusters_per_class=1, n_classes=3)
```

```python
plt.scatter(x[:,0] , x[:,1], s = 50, c=y)
```

```python
clf = SVC()
clf.fit(x,y) # trainning data
```

```python
plt.scatter(x[:,0] , x[:,1], s = 50, c = clf.predict(x)) # plot trained data
```

```python
#你有沒有看出哪個分錯了? 我是看不出來。所以我們用個簡單方式, 如果沒錯的會用一個顏色, 錯了就用其他顏色表示。
plt.scatter(x[:,0], x[:,1] ,s = 50 , c = clf.predict(x) - y)
```

```python
iris = load_iris()
```

```python
iris.target # the Y value
```

```python
x = iris.data[:,2:4] # pick 2 of the four for our x values
```

```python
len(x) == len(y)
```

```python
clf.fit(x,y)
```

```python
plt.scatter(x[:,0] , x[:,1], s = 100, c = y , marker = "*", cmap = plt.cm.rainbow_r, alpha = 0.5)
```
